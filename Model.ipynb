```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.svm import OneClassSVM
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from docx import Document
from docx.shared import Inches
import os

# Function to load and preprocess datasets
def load_and_preprocess(file_path):
    data = pd.read_csv(file_path, dtype=str).head(500)
    X = data[[' X-axis', ' Y-axis', ' Z-axis']]
    scaler = StandardScaler()
    X = scaler.fit_transform(X)
    return X

# List of paths to your datasets
dataset_paths = [
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\Healthy bearing data\Healthy with pulley.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\Healthy bearing data\Healthy without pulley.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\0.7mm-bearing-faults\0.7inner-100watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\0.7mm-bearing-faults\0.7inner-200watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\0.7mm-bearing-faults\0.7inner-300watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\0.7mm-bearing-faults\0.7outer-100watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\0.7mm-bearing-faults\0.7outer-200watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\0.7mm-bearing-faults\0.7outer-300watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\0.9mm-bearing-faults\0.9inner-100watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\0.9mm-bearing-faults\0.9inner-200watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\0.9mm-bearing-faults\0.9inner-300watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\0.9mm-bearing-faults\0.9outer-100watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\0.9mm-bearing-faults\0.9outer-200watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\0.9mm-bearing-faults\0.9outer-300watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.1mm-bearing-faults\1.1inner-100watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.1mm-bearing-faults\1.1inner-200watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.1mm-bearing-faults\1.1inner-300watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.1mm-bearing-faults\1.1outer-100watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.1mm-bearing-faults\1.1outer-200watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.1mm-bearing-faults\1.1outer-300watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.3mm-bearing-faults\1.3inner-100watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.3mm-bearing-faults\1.3inner-200watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.3mm-bearing-faults\1.3inner-300watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.3mm-bearing-faults\1.3outer-100watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.3mm-bearing-faults\1.3outer-200watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.3mm-bearing-faults\1.3outer-300watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.5mm-bearing-faults\1.5inner-100watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.5mm-bearing-faults\1.5inner-200watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.5mm-bearing-faults\1.5inner-300watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.5mm-bearing-faults\1.5outer-100watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.5mm-bearing-faults\1.5outer-200watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.5mm-bearing-faults\1.5outer-300watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.7mm-bearing-faults\1.7inner-100watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.7mm-bearing-faults\1.7inner-200watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.7mm-bearing-faults\1.7inner-300watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.7mm-bearing-faults\1.7outer-100watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.7mm-bearing-faults\1.7outer-200watt.csv",
    r"F:\Python\Motor Vibration Datat\fm6xzxnf36-2\1.7mm-bearing-faults\1.7outer-300watt.csv"
]
# Extract dataset names from paths
dataset_names = [os.path.basename(path).replace('.csv', '') for path in dataset_paths]

# Initialize lists to store metrics
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []

# Initialize variables to keep track of the best matching dataset
best_f1_score = 0
best_matching_dataset = None

# Define parameter grid for Grid Search
param_grid = {
    'nu': [0.01, 0.05, 0.1, 0.15, 0.2],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}

# Create a Word Document to save the results
doc_path = r"F:\Python\Motor Vibration Datat\Original Dataset\evaluation_main document.docx"
txt_path = r"F:\Python\Motor Vibration Datat\Original Dataset\evaluation_main_text.txt"

doc = Document()
doc.add_heading('Evaluation Results', 0)

with open(txt_path, 'w') as txt_file:
    for new_data_path in dataset_paths:
        new_data = pd.read_csv(new_data_path, dtype=str).head(500)
        new_data_name = os.path.basename(new_data_path)
        X_new = new_data[[' X-axis', ' Y-axis', ' Z-axis']]
        scaler = StandardScaler()
        X_new = scaler.fit_transform(X_new)
        
        f1_matrix = []
        
        for path in dataset_paths:
            X = load_and_preprocess(path)
            
            # Perform Grid Search for best parameters
            grid_search = GridSearchCV(OneClassSVM(), param_grid, scoring='f1', cv=5)
            grid_search.fit(X, [1] * len(X))  # Use [1] * len(X) assuming all are inliers
            
            # Get best model from Grid Search
            best_model = grid_search.best_estimator_
            
            # Predict on the new dataset
            y_pred = best_model.predict(X_new)
            
            # Convert predictions: -1 for outliers (anomalies), 1 for inliers (normal)
            y_pred = [0 if x == -1 else 1 for x in y_pred]
            
            # Assuming the new dataset is considered an anomaly (class 1)
            y_true = [1] * len(y_pred)
            
            # Calculate metrics
            accuracy = accuracy_score(y_true, y_pred)
            precision = precision_score(y_true, y_pred)
            recall = recall_score(y_true, y_pred)
            f1 = f1_score(y_true, y_pred)
            
            # Append metrics to the lists
            accuracy_scores.append(accuracy)
            precision_scores.append(precision)
            recall_scores.append(recall)
            f1_scores.append(f1)
            
            # Update the best matching dataset if the current F1 score is higher
            if f1 > best_f1_score:
                best_f1_score = f1
                best_matching_dataset = os.path.basename(path)
            
            f1_matrix.append(f1)
        
        # Create a DataFrame to display the results in a table format
        results_df = pd.DataFrame({
            'Dataset': dataset_names,
            'Accuracy': accuracy_scores,
            'Precision': precision_scores,
            'Recall': recall_scores,
            'F1-Score': f1_scores
        })
    
        # Rank the datasets based on F1-Score
        results_df['Rank'] = results_df['F1-Score'].rank(ascending=False, method='min')
        results_df = results_df.sort_values('Rank')
    
        # Write results to text file
        txt_file.write(f"Input Data name: {new_data_name}\n")
        txt_file.write("Results:\n")
        txt_file.write(results_df.to_string())
        txt_file.write("\n\nBest matching dataset based on F1-Score:\n")
        txt_file.write(best_matching_dataset)
        txt_file.write("\n")
        txt_file.write("#################################################################################\n")
        print(f"Input Data name: {new_data_name}\n")
        print("Results:\n")
        print(results_df.to_string())
        print("\n\nBest matching dataset based on F1-Score:\n")
        print(best_matching_dataset)
        print("\n")
        print("#################################################################################\n")
        
        # Add results to Word document
        doc.add_heading(f'Input Data name: {new_data_name}', level=1)
        doc.add_paragraph("Results:\n")
        
        # Add the table to the Word document
        table = doc.add_table(rows=1, cols=len(results_df.columns))
        hdr_cells = table.rows[0].cells
        for i, column in enumerate(results_df.columns):
            hdr_cells[i].text = column
        
        for _, row in results_df.iterrows():
            row_cells = table.add_row().cells
            for i, value in enumerate(row):
                row_cells[i].text = str(value)
        
        doc.add_paragraph("\n\nBest matching dataset based on F1-Score:\n")
        doc.add_paragraph(best_matching_dataset)
        doc.add_paragraph("#################################################################################\n")
        
        # Plot the line graph for F1 scores with ranks
        plt.figure(figsize=(12, 6))
        plt.plot(dataset_names, f1_matrix, marker='o', linestyle='-', color='b')
        plt.xticks(rotation=90)
        plt.xlabel('Datasets')
        plt.ylabel('F1 Score')
        plt.title(f'F1 Scores for {new_data_name}')
        
        # Highlight the dataset with the highest F1 score
        max_f1_index = f1_matrix.index(max(f1_matrix))
        plt.scatter([dataset_names[max_f1_index]], [f1_matrix[max_f1_index]], color='red', zorder=5)
        plt.annotate(f'Max F1: {max(f1_matrix):.5f}', 
                     xy=(dataset_names[max_f1_index], f1_matrix[max_f1_index]), 
                     xytext=(dataset_names[max_f1_index], f1_matrix[max_f1_index] + 0.02),
                     arrowprops=dict(facecolor='black', shrink=0.05))
        plt.gca().get_xticklabels()[max_f1_index].set_color('red')
        
        # Save the plot to a file
        plot_path = r"F:\Python\Motor Vibration Datat\Original Dataset\plot_{new_data_name}.png"
        plt.tight_layout()
        plt.savefig(plot_path)
        plt.show()
        
        # Add the plot to the Word document
        doc.add_heading(f'F1 Scores for {new_data_name}', level=2)
        doc.add_picture(plot_path, width=Inches(6))
        
        # Reset metrics lists for the next iteration
        accuracy_scores = []
        precision_scores = []
        recall_scores = []
        f1_scores = []
        best_f1_score = 0
        best_matching_dataset = None

# Save the Word document
doc.save(doc_path)

print(f'Results have been saved to {doc_path} and {txt_path}')
```
